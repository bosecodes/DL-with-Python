```python
# imputation of some missing data
```

```python
plt.figure(figsize = (10,7))
sns.boxplot(x = 'Pclass', y = 'Age', data = train)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age, axis = 1)
```

```python
sns.heatmap(train.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
train.drop('Cabin', axis = 1, inplace = True)
```

```python
train.head()
```

```python
train.dropna(inplace = True)
```

```python
sns.heatmap(train.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
sex = pd.get_dummies(train['Sex'], drop_first = True)
```

```python
embark = pd.get_dummies(train['Embarked'], drop_first= True)
```

```python
embark.head()
```

```python
train = pd.concat([train, sex, embark], axis = 1)
```

```python
train.head()
```

```python
train.drop(['Name', 'Ticket', 'Sex', 'Embarked'], axis = 1, inplace = True)
```

```python
train.drop('PassengerId', axis = 1, inplace = True)
```

```python
train.head()
```

```python
first = pd.get_dummies(train['Pclass'], drop_first= True)
```

```python
first.head()
```

```python
train = pd.concat([train, first], axis = 1)
```

```python
train.drop('Pclass', axis = 1, inplace = True)
```

```python
train.head()
```

```python
x = train.drop('Survived', axis = 1)
y =  train['Survived'] #col that you wanna predict
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train, y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
#for classification, scikitlearn has a good feature

from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```
